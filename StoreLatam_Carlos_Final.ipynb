### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/cbermell/StoreLatamDecision/refs/heads/main/base-de-datos-latam/tienda_1.csv"
url2 = "https://raw.githubusercontent.com/cbermell/StoreLatamDecision/refs/heads/main/base-de-datos-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/cbermell/StoreLatamDecision/refs/heads/main/base-de-datos-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/cbermell/StoreLatamDecision/refs/heads/main/base-de-datos-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
""""
Ingreso total por cada tienda:
En este primer análisis, calculamos el ingreso total de cada tienda.
Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.
"""
```

```python
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()
```

```python
print(f"Ingreso total Tienda 1: ${ingreso_tienda_1:.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda_2:.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda_3:.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda_4:.2f}")
```

# 2. Ventas por categoría

```python
"""
Ventas por categoría:
En este calculamos la cantidad de productos vendidos por categoría en cada tienda.
La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo,
mostrando las categorías más populares de cada tienda.
"""
```

```python
ventas_categoria_1 = tienda['Categoría del Producto'].value_counts()
print("Ventas por categoría - Tienda 1")
print(ventas_categoria_1)
print("\n")
ventas_categoria_2 = tienda2['Categoría del Producto'].value_counts()
print("Ventas por categoría - Tienda 2")
print(ventas_categoria_2)
print("\n")
ventas_categoria_3 = tienda3['Categoría del Producto'].value_counts()
print("Ventas por categoría - Tienda 3")
print(ventas_categoria_3)
print("\n")
ventas_categoria_4 = tienda4['Categoría del Producto'].value_counts()
print("Ventas por categoría - Tienda 4")
print(ventas_categoria_4)
```

# 3. Calificación promedio de la tienda


```python
"""
Valoración media por tienda:
Aqui calculamos las calificaciones promedio de los clientes para cada tienda.
El objetivo es conocer la satisfacción del cliente con los productos vendidos.
"""
```

```python
media_1 = tienda["Calificación"].mean()

media_2 = tienda2["Calificación"].mean()

media_3 = tienda3["Calificación"].mean()

media_4 = tienda4["Calificación"].mean()

print(f"Valoración media de la Tienda 1: {media_1:.2f}")
print(f"Valoración media de la Tienda 2: {media_2:.2f}")
print(f"Valoración media de la Tienda 3: {media_3:.2f}")
print(f"Valoración media de la Tienda 4: {media_4:.2f}")
```

# 4. Productos más y menos vendidos

```python
ventas_productos = tienda['Producto'].value_counts()

producto_top = ventas_productos.idxmax()
ventas_top = ventas_productos.max()

producto_menos = ventas_productos.idxmin()
ventas_menos = ventas_productos.min()

print("Producto más vendido:", producto_top, "-", ventas_top, "ventas")
print("Producto menos vendido:", producto_menos, "-", ventas_menos, "ventas")
```

# 5. Envío promedio por tienda

```python
promedio_envio1 = tienda["Costo de envío"].mean()
print(f"Costo de envío promedio - Tienda 1 :{promedio_envio1:.2f}")

promedio_envio2 = tienda2["Costo de envío"].mean()
print(f"Costo de envío promedio - Tienda 2 :{promedio_envio2:.2f}")

promedio_envio3 = tienda3["Costo de envío"].mean()
print(f"Costo de envío promedio - Tienda 3 :{promedio_envio3:.2f}")

promedio_envio4 = tienda4["Costo de envío"].mean()
print(f"Costo de envío promedio - Tienda 4 :{promedio_envio4:.2f}")
```

#INFORME FINAL
El presente informe tiene como objetivo determinar cuál de las cuatro tiendas analizadas representa la mejor opción para que el Sr. Juan comercialice sus productos. A través del análisis de datos reales proporcionados por cada tienda, se han evaluado distintos aspectos clave: ingresos totales, categorías de productos más y menos vendidas, calificaciones promedio de los clientes, productos más vendidos y el costo de envío promedio.
Estos factores son esenciales para tomar una decisión informada que maximice las posibilidades de éxito comercial del Sr. Juan.



#Ingresos Totales
El primer análisis reveló que:

La tienda 3 obtuvo el mayor ingreso total, seguida por la Tienda 2.

Las tiendas 1 y 4 generaron ingresos más bajos en comparación.

Este dato sugiere que la Tienda 3 tiene un mayor volumen de ventas o vende productos de mayor valor, lo que puede representar una mayor oportunidad de ganancias para el Sr. Juan.

#Gráfico - Ingresos totales por tienda




```python
ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
tiendas = ["T1", "T2", "T3", "T4"]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='royalblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')

plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, ingreso in enumerate(ingresos):
    plt.text(i, ingreso + 0.005, f'${ingreso:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

#Ventas por Categoría
En cuanto a la popularidad de las categorías de productos:

En la Tienda 1, las categorías más vendidas fueron “Tecnología” y “Moda”.

Algunas categorías mostraron muy bajo volumen, como “Hogar” o “Juguetes”.

Esto permite al Sr. Juan alinear su oferta con las preferencias de cada tienda. Si sus productos pertenecen a categorías populares, será más fácil posicionarlos.



```python
colores = ['royalblue', 'darkorange', 'seagreen', 'crimson']
tiendas = [ventas_categoria_1, ventas_categoria_2, ventas_categoria_3, ventas_categoria_4]
urls = [url, url2, url3, url4]

plt.figure(figsize=(12,6))

for i, url in enumerate(urls):
    df = pd.read_csv(url)
    tiendas = df['Categoría del Producto'].value_counts()
    categorias = tiendas.index
    cantidades = tiendas.values

    plt.scatter(categorias, cantidades, color=colores[i],s=100, label=tiendas[i],)

plt.title("Ventas por categoría en cada tienda (Dispersión)")
plt.xlabel("Categorías")
plt.ylabel("Cantidad de productos vendidos")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title="Tiendas")
plt.tight_layout()
plt.show()
```

#Gráfico - Valoración promedio por tienda

```python
valoraciones = [media_1, media_2, media_3, media_4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8,5))
plt.bar(tiendas, valoraciones, color='coral')
plt.title("Valoración promedio por tienda")
plt.ylabel("Valoración promedio")
plt.ylim(3.6, 4.4)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, val in enumerate(valoraciones):
    plt.text(i, val + 0.05, f"{val:.2f}", ha='center')

plt.tight_layout()
plt.show()
```

#Costo de Envío Promedio
Los promedios de costo de envío fueron:

#Gráfico - Costo de envio promedio por tienda


```python
costos_envio = [promedio_envio1, promedio_envio2, promedio_envio3, promedio_envio4]

plt.figure(figsize=(8,5))
plt.scatter(tiendas, costos_envio, color='crimson', s=100)  # s es el tamaño del punto
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo promedio ($)")
plt.grid(True, linestyle='--', alpha=0.6)

for i, costo in enumerate(costos_envio):
    plt.text(i, costo + 0.02, f"${costo:.2f}", ha='center')

plt.tight_layout()
plt.show()
```
